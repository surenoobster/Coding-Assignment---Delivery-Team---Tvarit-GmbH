```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/data.csv')

# Drop the timestamp column
df = df.drop(columns=['timestamp'])

# Define a function to join consecutive rows
def join_consecutive_rows(df):
    new_rows = []
    for i in range(0, len(df), 4):
        chunk = df.iloc[i:i+4].reset_index(drop=True)
        new_row = {}
        for j, (_, row) in enumerate(chunk.iterrows()):
            for col in df.columns[:-1]:  # Exclude the label column
                new_row[f'{col}_{chr(97+j)}'] = row[col]
            new_row[f'label_{chr(97+j)}'] = row['label']
        new_rows.append(new_row)
    return pd.DataFrame(new_rows)

# Apply the function to the DataFrame
result_df = join_consecutive_rows(df)

# Print the result
print(result_df)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/windowedmodified.csv')  # Replace with the path to your dataset

# Define the columns to calculate max, min, and standard deviation
columns_to_calculate = [
    ['temperature_1_a', 'temperature_1_b', 'temperature_1_c', 'temperature_1_d'],
    ['temperature_2_a', 'temperature_2_b', 'temperature_2_c', 'temperature_2_d'],
    ['temperature_3_a', 'temperature_3_b', 'temperature_3_c', 'temperature_3_d'],
    ['temperature_4_a', 'temperature_4_b', 'temperature_4_c', 'temperature_4_d']
]

# Initialize new columns
for prefix in ['temp_max', 'temp_min', 'temp_std']:
    for i in range(1, 5):
        df[f'{prefix}_{i}'] = None

# Perform calculations row by row
for i, row in df.iterrows():
    for j, cols in enumerate(columns_to_calculate):
        # Extract values for the current row and columns
        values = row[cols]
        # Calculate max, min, and standard deviation
        max_val = values.max()
        min_val = values.min()
        std_dev = values.std()
        # Update the new columns
        df.at[i, f'temp_max_{j+1}'] = max_val
        df.at[i, f'temp_min_{j+1}'] = min_val
        df.at[i, f'temp_std_{j+1}'] = std_dev

# Print the updated DataFrame
print(df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Load the filtered dataset
dataset_path = "/content/filtered_data.csv"
data = pd.read_csv(dataset_path)

# Split features (X) and target variable (y)
X = data.drop(columns=['label_d'])  # Features
y = data['label_d']  # Target variable

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(max_iter=1000, penalty='l2', C=1.0),
    'SVM (RBF Kernel)': SVC(kernel='rbf', C=1.0),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, min_samples_leaf=1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, max_depth=3),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5)
}

# Train and evaluate each classifier using cross-validation
for clf_name, clf in classifiers.items():
    # Perform cross-validation
    cv_scores = cross_val_score(clf, X_scaled, y, cv=5)
    avg_accuracy = cv_scores.mean()
    print(f"{clf_name} Cross-Validation Accuracy: {avg_accuracy:.4f}")

    # Fit the classifier on the entire training set
    clf.fit(X_train, y_train)

    # Predict on the test set
    y_pred = clf.predict(X_test)

    # Calculate additional metrics
    report = classification_report(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)

    # Print additional metrics
    print(f"\n{clf_name} Classification Report:")
    print(report)
    print(f"\n{clf_name} ROC AUC Score: {roc_auc:.4f}\n")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE

# Load the filtered dataset
dataset_path = "/content/filtered_data.csv"
data = pd.read_csv(dataset_path)

# Split features (X) and target variable (y)
X = data.drop(columns=['label_d'])  # Features
y = data['label_d']  # Target variable

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Initialize classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(max_iter=1000, penalty='l2', C=1.0),
    'SVM (RBF Kernel)': SVC(kernel='rbf', C=1.0),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, min_samples_leaf=1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, max_depth=3),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5)
}

# Train and evaluate each classifier using cross-validation
for clf_name, clf in classifiers.items():
    # Perform cross-validation
    cv_scores = cross_val_score(clf, X_resampled, y_resampled, cv=5)
    avg_accuracy = cv_scores.mean()
    print(f"{clf_name} Cross-Validation Accuracy: {avg_accuracy:.4f}")

    # Fit the classifier on the entire training set
    clf.fit(X_train, y_train)

    # Predict on the test set
    y_pred = clf.predict(X_test)

    # Calculate additional metrics
    report = classification_report(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)

    # Print additional metrics
    print(f"\n{clf_name} Classification Report:")
    print(report)
    print(f"\n{clf_name} ROC AUC Score: {roc_auc:.4f}\n")
```
